##### 放在一起

```python
from tqdm import tqdm
def data_cleaner(data):
    clean_data = []
    for review in tqdm(data['text']):
        cleantext = BeautifulSoup(review, "lxml").text
        cleantext = re.sub(r'[^\w\s]', '', cleantext)
        cleantext = [ token for token in cleantext.lower().split() if token not in stopword ]
        cleantext = " ".join([lemmatizer.lemmatize(t) for t in cleantext])
        clean_data.append(cleantext.strip())
    return clean_data
```

```python
clean_train = data_cleaner(data_train)
```

## 特征提取

```python
from nltk.corpus import wordnet as wn
from nltk.corpus import sentiwordnet as swn
from nltk.tokenize import word_tokenize 
```

```python
token_list = word_tokenize(clean_train[0])
pos_tags = nltk.pos_tag(token_list)
```

```python
def penn_to_wn(tag):
    if tag.startswith('J'):
        return wn.ADJ
    elif tag.startswith('N'):
        return wn.NOUN
    elif tag.startswith('R'):
        return wn.ADV
    elif tag.startswith('V'):
        return wn.VERB
    return None
```

```python
pos_tags
```

```python
synsets = wn.synsets(token_list[4], pos=penn_to_wn(pos_tags[4][0]))
synsets
```

```python
swn_entry = swn.senti_synset(synsets[0].name())
swn_entry
```

```python
synsets[0].name(), swn_entry.pos_score(),swn_entry.neg_score()
```
